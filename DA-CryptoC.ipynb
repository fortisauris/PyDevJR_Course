```python
import pandas as pd
import matplotlib.pyplot as mp
```

```python
btcusd_url = 'https://www.cryptodatadownload.com/cdd/Bitfinex_BTCUSD_1h.csv'
content = pd.read_csv(btcusd_url, skiprows=1)
content.drop(['unix','Volume USD', 'Volume BTC'], axis=1, inplace=True)
print(content.head(10))
print(content.columns)
content.shape
```

```python
mp.figure(figsize=(15,5))
data = content.drop(content.index[24:])
data = data[::-1]
mp.plot(data['date'].str[10:13], data['close'], color='blue')
mp.plot(data['date'].str[10:13], data['open'])
mp.plot(data['date'].str[10:13], data['low'])
mp.plot(data['date'].str[10:13], data['high'])
mp.title('BTC/USD za poslednych 24h')
mp.show()
```

```python
btceur_url = 'https://www.cryptodatadownload.com/cdd/Bitfinex_BTCEUR_d.csv'
content = pd.read_csv(btceur_url, skiprows=1)
content.drop(['unix','Volume EUR', 'Volume BTC'], axis=1, inplace=True)
print(content.head(5))
content.shape
```

```python
mp.figure(figsize=(15,5))
data = content.drop(content.index[31:])
data = data[::-1]
mp.xticks(ticks=range(len(data)), labels=data['date'].str[:10], rotation=90)
mp.plot(data['date'], data['open'], color='blue')
mp.plot(data['date'], data['close'])
mp.plot(data['date'], data['low'])
mp.plot(data['date'], data['high'])
# print(data)
mp.title('BTC/EUR za poslednych 31 dni')
mp.show()
```
