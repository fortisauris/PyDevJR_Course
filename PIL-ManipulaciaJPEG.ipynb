```python
from PIL import Image, ImageEnhance, ImageFont, ImageDraw
```

```python
im = Image.open('coder.jpg')
print(im.format, im.size)
# im.show()
```

```python
display(im)
```

```python
#x = int(im.size[0] * 1.5)
#y = int(im.size[1] * 1.5)
#im = im.resize((x,y))
print(im.size, im.mode)  # R G B - RED GREEN BLUE
display(im)
```

```python
r,g,b = im.split()
display(r,g,b)
```

```python
im = Image.merge('RGB', (r,g,b))
display(im)
```

```python
im = im.transpose(Image.Transpose.FLIP_LEFT_RIGHT)
display(im)
```

```python
display(im.convert('L'))
```

```python
out = im.point(lambda i: i * 0.5)  # Image Contrast Manipulation
display(out)
```

```python
enhanced = ImageEnhance.Contrast(im)
display(enhanced.enhance(1.5))
enhanced = ImageEnhance.Color(im)
display(enhanced.enhance(2))  # pravdepodobne Saturacia farieb
enhanced = ImageEnhance.Brightness(im)
display(enhanced.enhance(2))  # JAS
enhanced = ImageEnhance.Sharpness(im)
display(enhanced.enhance(1.5))  # OSTROST
```

```python
box = (50,50, 150,150)
cast_obrazku = im.crop(box)
display(cast_obrazku)
cast_obrazku.size
```

```python
cast_obrazku = cast_obrazku.resize((100,100))
display(cast_obrazku)
display(im)
```

```python
x = int(im.size[0] * 1.5)
y = int(im.size[1] * 1.5)
# im = im.resize((x,y))
box = (300,300,400,400)
im.paste(cast_obrazku, box)
display(im)
```
